```python
import pandas as pd

# It's actually a CSV file, not real Excel
df = pd.read_csv("data/Foreign_Exchange_Rates (2).xls")

print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df.head()
```

```python
# Drop unnamed/empty columns
df = df.drop(columns=["Unnamed: 0", "Unnamed: 24"], errors="ignore")

# Rename Time Serie to Date
df = df.rename(columns={"Time Serie": "Date"})

# Convert Date - format is DD-MM-YYYY
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y", errors="coerce")

print("Missing dates:", df["Date"].isna().sum())

# Sort by date
df = df.sort_values("Date").reset_index(drop=True)

# Convert currency columns to numeric (catches 'ND' or text as missing)
for col in df.columns[1:]:
    df[col] = pd.to_numeric(df[col], errors="coerce")

print("\nMissing values per column:")
print(df.isna().sum())

print("\nShape after cleaning:", df.shape)
df.head()
```

```python
# Forward-fill missing values (use previous day's rate)
# This makes sense for time-series financial data
df = df.fillna(method="ffill")

# In case the very first rows have missing data, backfill too
df = df.fillna(method="bfill")

print("Missing values after fill:")
print(df.isna().sum().sum(), "total missing values remaining")

print("\nFinal shape:", df.shape)
print("\nDate range:", df["Date"].min(), "to", df["Date"].max())
df.head()
```

```python
model = Prophet(
    daily_seasonality=False,
    yearly_seasonality=False,
    weekly_seasonality=False,
    changepoint_prior_scale=0.5  # allows model to adapt to trend changes better
)
model.fit(train)

print("Model trained successfully!")
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Create future dataframe for the test period
future = model.make_future_dataframe(periods=60)
forecast = model.predict(future)

# Get predictions for the test period only
predictions = forecast.iloc[-60:]["yhat"].values
actual = test["y"].values

# Calculate metrics
mae = mean_absolute_error(actual, predictions)
rmse = np.sqrt(mean_squared_error(actual, predictions))
mape = np.mean(np.abs((actual - predictions) / actual)) * 100

print(f"MAE:  {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAPE: {mape:.2f}%")
```

```python
import warnings
warnings.filterwarnings("ignore")

results = []

currency_cols = df.columns[1:]  # all columns except Date

def evaluate_naive(train_y, test_y):
    pred = np.full(len(test_y), train_y.iloc[-1])
    mae = mean_absolute_error(test_y, pred)
    rmse = np.sqrt(mean_squared_error(test_y, pred))
    mape = np.mean(np.abs((test_y - pred) / test_y)) * 100
    return mae, rmse, mape

def evaluate_prophet(train_df, test_df):
    m = Prophet(
        daily_seasonality=False,
        yearly_seasonality=False,
        weekly_seasonality=False,
        changepoint_prior_scale=0.01
    )
    m.fit(train_df)
    future = m.make_future_dataframe(periods=len(test_df))
    forecast = m.predict(future)
    pred = forecast.iloc[-len(test_df):]["yhat"].values
    actual = test_df["y"].values
    mae = mean_absolute_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mape = np.mean(np.abs((actual - pred) / actual)) * 100
    return mae, rmse, mape

# Test on first 3 currencies to confirm it works before running all 22
for currency in currency_cols[:3]:
    print(f"Processing: {currency}")

    cur_df = df[["Date", currency]].copy()
    cur_df.columns = ["ds", "y"]

    train = cur_df.iloc[:-60]
    test = cur_df.iloc[-60:]

    naive_mae, naive_rmse, naive_mape = evaluate_naive(train["y"], test["y"])
    prophet_mae, prophet_rmse, prophet_mape = evaluate_prophet(train, test)

    results.append({
        "currency": currency,
        "naive_mae": naive_mae, "naive_rmse": naive_rmse, "naive_mape": naive_mape,
        "prophet_mae": prophet_mae, "prophet_rmse": prophet_rmse, "prophet_mape": prophet_mape,
        "best_model": "naive" if naive_mae < prophet_mae else "prophet"
    })

results_df = pd.DataFrame(results)
print(results_df)
```

```python
from xgboost import XGBRegressor

def create_features(series, n_lags=5):
    """Create lag features for XGBoost"""
    data = pd.DataFrame({"y": series.values})
    for lag in range(1, n_lags + 1):
        data[f"lag_{lag}"] = data["y"].shift(lag)
    data = data.dropna().reset_index(drop=True)
    return data

def evaluate_xgboost(train_y, test_y, n_lags=5):
    full_series = pd.concat([train_y, test_y]).reset_index(drop=True)
    feat_df = create_features(full_series, n_lags)

    split_point = len(train_y) - n_lags
    train_feat = feat_df.iloc[:split_point]
    test_feat = feat_df.iloc[split_point:split_point + len(test_y)]

    X_train = train_feat.drop(columns=["y"])
    y_train = train_feat["y"]
    X_test = test_feat.drop(columns=["y"])
    y_test = test_feat["y"]

    model = XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.05)
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, pred)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    mape = np.mean(np.abs((y_test - pred) / y_test)) * 100
    return mae, rmse, mape

# Test on EUR first
eur_df2 = df[["Date", "EURO AREA - EURO/US$"]].copy()
eur_df2.columns = ["ds", "y"]
train2 = eur_df2.iloc[:-60]
test2 = eur_df2.iloc[-60:]

xgb_mae, xgb_rmse, xgb_mape = evaluate_xgboost(train2["y"], test2["y"])
print(f"XGBoost - MAE: {xgb_mae:.4f}, RMSE: {xgb_rmse:.4f}, MAPE: {xgb_mape:.2f}%")
```

```python
import warnings
warnings.filterwarnings("ignore")

results = []
currency_cols = df.columns[1:]

for currency in currency_cols:
    print(f"Processing: {currency}")

    cur_df = df[["Date", currency]].copy()
    cur_df.columns = ["ds", "y"]

    train = cur_df.iloc[:-60]
    test = cur_df.iloc[-60:]

    naive_mae, naive_rmse, naive_mape = evaluate_naive(train["y"], test["y"])
    prophet_mae, prophet_rmse, prophet_mape = evaluate_prophet(train, test)
    xgb_mae, xgb_rmse, xgb_mape = evaluate_xgboost(train["y"], test["y"])

    scores = {"naive": naive_mae, "prophet": prophet_mae, "xgboost": xgb_mae}
    best_model = min(scores, key=scores.get)

    results.append({
        "currency": currency,
        "naive_mape": round(naive_mape, 2),
        "prophet_mape": round(prophet_mape, 2),
        "xgboost_mape": round(xgb_mape, 2),
        "best_model": best_model
    })

results_df = pd.DataFrame(results)
print("\n=== FINAL COMPARISON ===")
print(results_df.to_string(index=False))
```

```python
import joblib
import os

os.makedirs("models", exist_ok=True)

def train_final_xgboost(series, n_lags=5):
    """Train XGBoost on full data using lag features"""
    data = pd.DataFrame({"y": series.values})
    for lag in range(1, n_lags + 1):
        data[f"lag_{lag}"] = data["y"].shift(lag)
    data = data.dropna().reset_index(drop=True)

    X = data.drop(columns=["y"])
    y = data["y"]

    model = XGBRegressor(n_estimators=100, max_depth=3, learning_rate=0.05)
    model.fit(X, y)
    return model

# Map currency column names to simple codes for filenames
currency_code_map = {
    "AUSTRALIA - AUSTRALIAN DOLLAR/US$": "AUD",
    "EURO AREA - EURO/US$": "EUR",
    "NEW ZEALAND - NEW ZELAND DOLLAR/US$": "NZD",
    "UNITED KINGDOM - UNITED KINGDOM POUND/US$": "GBP",
    "BRAZIL - REAL/US$": "BRL",
    "CANADA - CANADIAN DOLLAR/US$": "CAD",
    "CHINA - YUAN/US$": "CNY",
    "HONG KONG - HONG KONG DOLLAR/US$": "HKD",
    "INDIA - INDIAN RUPEE/US$": "INR",
    "KOREA - WON/US$": "KRW",
    "MEXICO - MEXICAN PESO/US$": "MXN",
    "SOUTH AFRICA - RAND/US$": "ZAR",
    "SINGAPORE - SINGAPORE DOLLAR/US$": "SGD",
    "DENMARK - DANISH KRONE/US$": "DKK",
    "JAPAN - YEN/US$": "JPY",
    "MALAYSIA - RINGGIT/US$": "MYR",
    "NORWAY - NORWEGIAN KRONE/US$": "NOK",
    "SWEDEN - KRONA/US$": "SEK",
    "SRI LANKA - SRI LANKAN RUPEE/US$": "LKR",
    "SWITZERLAND - FRANC/US$": "CHF",
    "TAIWAN - NEW TAIWAN DOLLAR/US$": "TWD",
    "THAILAND - BAHT/US$": "THB"
}

saved_models = []

for currency_col, code in currency_code_map.items():
    print(f"Training final model for {code}...")
    model = train_final_xgboost(df[currency_col])

    filepath = f"models/{code}_model.pkl"
    joblib.dump(model, filepath)
    saved_models.append(code)

print(f"\nAll {len(saved_models)} models saved successfully!")
print("Saved models:", saved_models)
```
